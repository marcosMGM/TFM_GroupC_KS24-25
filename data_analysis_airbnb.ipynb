Vamos a sumar todos los "host_listings_count" por cada uno de los propietarios para ver si coincide con "host_total_listings_count".

```python
# Agrupamos el dataframe por los host_id y sumamos todos los valores por cada host
host_total_listing_count_m = data.groupby('host_id').sum()
# Creamos una serie donde aparezca la suma de host_listings_count por cada uno de los host_id
host_listings_count_sum = pd.Series(host_total_listing_count_m['host_listings_count'], index=host_total_listing_count_m.index, name='total_listings_count')
# Sacamos del dataframe original las columnas de "host_id" y "host_total_listings_count"
comparison_ds = data[['host_id', 'host_total_listings_count']]
# Añadimos al anterior dataframe la suma de "host_listings_count" por cada uno de los "host_id" y comparamos si coinciden con los vamores de "host_total_listings_count"
comparison_ds['total_listings_count'] = comparison_ds['host_id'].map(host_listings_count_sum)
comparison_ds
```

No coinciden.

Vamos a comparar ambos valores ("host_listings_count" y "host_total_listings_count) ordenándolo por "host_id".

```python
listings = data[['host_id', 'host_listings_count', 'host_total_listings_count']].sort_values(by='host_id')
listings.head(50)
```

Parece que "host_listings_count" siempre es igual o menor a "host_total_listings_count". Esto puede decir que el primero hace referencia a los anuncios que tiene cada host activos y el segundo, a todos los anuncios, estén o no públicos.

Es posible que, cuantos más anuncios tenga el host en AirBnB, sus apartamentos por separado tengan más reservas al tener los usuarios más referencias sobre el propietario.

```python
import matplotlib.pyplot as plt
```

```python
scatter = plt.scatter(
    x = data['estimated_occupancy_l365d'],
    y = data['host_listings_count'],
    c = data['host_id'],
    cmap = 'viridis',
    alpha = 0.3
)
plt.colorbar(scatter, label = 'host_id')
plt.xlabel('estaimated_occupancy_l365d')
plt.ylabel('host_listings_count')
plt.ylim(0, 1100)
```

```python
scatter = plt.scatter(
    x = data['estimated_occupancy_l365d'],
    y = data['host_total_listings_count'],
    c = data['host_id'],
    cmap = 'viridis',
    alpha = 0.3
)
plt.colorbar(scatter, label = 'host_id')
plt.xlabel('estaimated_occupancy_l365d')
plt.ylabel('host_total_listings_count')
plt.ylim(0, 1100)
```

No da la sensación de que la ocupación sea mayor si el host tiene más anuncios. Es más, parece que aquellos que tienen excesivos anuncios llegan a tener mucha menos ocupación. Posible redundancia de anuncios para un mismo apartamento.
