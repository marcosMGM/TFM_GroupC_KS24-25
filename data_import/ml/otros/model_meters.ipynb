```python
# División de los datos en train y test
X = df_houses[["HABITACIONES", "BAÑOS", "JARDIN",  "TERRAZA"]]
y = df_houses["SUPERFICIE_M2"]

X_train, X_test, y_train, y_test = train_test_split(X, y, train_size = 0.8, random_state = 42)
```

```python
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

X_train = pd.DataFrame(X_train_scaled, columns=X_train.columns, index=X_train.index)
```

```python
X_train
```

```python
X_train = sm.add_constant(X_train, prepend=True) #Para poder calcular el valor de las constantes
model_linear = sm.OLS(endog=y_train, exog=X_train)
model_linear = model_linear.fit()
```

```python
print(model_linear.summary())
```

```python
# Error de test del modelo 
X_test = sm.add_constant(X_test_scaled, prepend=True)
predic = model_linear.predict(exog = X_test)

r_2 = r2_score(
        y_true  = y_test,
        y_pred  = predic)

#print(f"El error (rmse) de test es: {rmse}")
print(f"El r^2 de test es: {r_2}")
```
