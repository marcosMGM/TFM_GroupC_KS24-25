```python
TIPOS_TRANSPORTE = [4,5,6,8,10] #El 9 está integrado en el 8
```

```python
def calcular_distancias_vivienda(row_viv):
    resultados = {}
    
    for modo in TIPOS_TRANSPORTE:
        paradas_tipo = stops[stops['mode'] == modo]

        if paradas_tipo.empty:
            resultados[f'distancia_mode_{modo}'] = None
            continue

        delta = 0.01
        paradas_candidatas = paradas_tipo[
            (abs(paradas_tipo['lat'] - row_viv['LATITUDE']) < delta) &
            (abs(paradas_tipo['lon'] - row_viv['LONGITUDE']) < delta)
        ]


        """ Para mejorar el rendimiento voy iterando incrementando el delta a comprobar, tengo que 
        asegurarme de que haya al menos una parada en el dataset porque sino morimos en el while"""
        while paradas_candidatas.empty:  
            delta += 0.01
            print(f"Buscando paradas cercanas al modo {modo} con delta {delta}")
            paradas_candidatas = paradas_tipo[
                (abs(paradas_tipo['lat'] - row_viv['LATITUDE']) < delta) &
                (abs(paradas_tipo['lon'] - row_viv['LONGITUDE']) < delta)
            ]

        # if paradas_candidatas.empty:
        #     paradas_candidatas = paradas_tipo  # fallback a todas

        distancias = paradas_candidatas.apply(
            lambda stop: distance(
                (row_viv['LATITUDE'], row_viv['LONGITUDE']),
                (stop['lat'], stop['lon'])
            ).meters,
            axis=1
        )

        resultados[f'distancia_mode_{modo}'] = distancias.min()

    return pd.Series(resultados)

```

```python
for index, row in df.iterrows():
    house_id = row['HOUSE_ID']
    distancia_metro   = row.get('distancia_mode_4', 'NULL')
    distancia_cercanias   = row.get('distancia_mode_5', 'NULL')
    distancia_emt   = row.get('distancia_mode_6', 'NULL')
    distancia_bus  = row.get('distancia_mode_8', 'NULL')
    distancia_mlo = row.get('distancia_mode_10', 'NULL')

    distancia_metro = 'NULL' if pd.isna(distancia_metro) else round(distancia_metro, 0)
    distancia_cercanias  = 'NULL' if pd.isna(distancia_cercanias) else round(distancia_cercanias, 0)
    distancia_emt   = 'NULL' if pd.isna(distancia_emt) else round(distancia_emt, 0)
    distancia_bus   = 'NULL' if pd.isna(distancia_bus) else round(distancia_bus, 0)
    distancia_mlo   = 'NULL' if pd.isna(distancia_mlo) else round(distancia_mlo, 0)

    sql = f"""
    UPDATE HOUSES
    SET DISTANCE_TO_METRO = {distancia_metro},
    DISTANCE_TO_CERCANIAS = {distancia_cercanias},
    DISTANCE_TO_EMT = {distancia_emt},
    DISTANCE_TO_INTERURBANOS = {distancia_bus},
    DISTANCE_TO_MLO = {distancia_mlo}
    WHERE HOUSE_ID = {house_id};
    """
    cursor = conn.cursor()
    cursor.execute(sql)
    conn.commit()
```
